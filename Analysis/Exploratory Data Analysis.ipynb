```python
def hist_plot(dat, col, facecolor = "green", normed = 1, scale_bins = 4):
    num_bins = int(dat.groupby(col).count().shape[0]/scale_bins)
    plot_for_output = plt.hist(dat[col],facecolor = "green", normed = normed,bins = num_bins)
    return(plot_for_output)

```

```python
hold = hist_plot(dat = df,col = "SalePrice")
```
